```python
new_X = new.drop(['Label'],axis=1)
y = new['Label']
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
```

```python
std.fit(new_X)

# 학습 데이터 스케일링
X_scaled = std.transform(new_X)
X = X_scaled
```

```python
X = pd.DataFrame(X)
```

```python
def Decision_rate(a):
   # print("threshold a is : ", a)
    full_acc = 0
    for j in range(5):
        print("")
        print("=======================================")
     #   print("Column index: ", j)
     #   print("=======================================")
        
        TP = 0
        for i in range(39614):
            if X[j][i] >= a and y[i] == 1:
                TP += 1
     #   print("(TP) attack >= ", TP)

        FN = 0
        for i in range(39614):
            if X[j][i] < a and y[i] == 1:
                FN += 1
     #   print("(FN) attack < ", FN)

        TN = 0
        for i in range(39614):
            if X[j][i] >= a and y[i] == 0:
                TN += 1
     #   print("(TN) Benign >= ", TN)


        FP = 0
        for i in range(39614):
            if X[j][i] < a and y[i] == 0:
                FP += 1
     #   print("(FP) Benign < ", FP)
        
        
        #performance
        accuracy = (TP + TN) / (TP + TN + FP + FN)
        precision = TP / (TP + FP)
        
        print("Accuracy: ", accuracy, "      Precision: ", precision)
        full_acc += accuracy
    full_acc /= 4
    return full_acc
```
