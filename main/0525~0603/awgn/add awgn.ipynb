```python
import numpy as np
```

```python
mu, sigma = 0, 0.1
```

```python
noise = np.random.normal(mu, sigma, [11884,27])
```

```python
import pandas as pd
```

```python
without = pd.read_csv("../sampling_dataset/CIRA_50%_train.csv")
```

```python
X = without.drop(['Unnamed: 0','Label'], axis=1)
X
```

```python
y = without['Label']
y
```

```python
noise_X = X+noise
noise_X
```

```python
noise_XY = pd.concat([noise_X, y], axis=1)
noise_XY
```

```python
noise_XY.to_csv("./WITH sampling/CIRA_50%_train_noise.csv")
```
