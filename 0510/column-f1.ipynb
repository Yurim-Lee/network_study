```python
def performance(prediction):
    TP = 0
    for i in range(7923):
        if prediction[i] == label_list[i][0] ==1:
            TP+=1
    
    FN = 0
    for i in range(7923):
        if prediction[i] == 0 and label_list[i][0] == 1:
            FN += 1
    
    TN = 0
    for i in range(7923):
        if prediction[i] == label_list[i][0] == 0:
            TN += 1
    
    FP = 0
    for i in range(7923):
        if prediction[i] == 1 and label_list[i][0] ==0:
            FP += 1
    return TP, FN, TN, FP      
        
```

```python
def performance_result(TP, FN, TN, FP):
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    print("accuracy: ", accuracy)
    
    precision = TP / (TP + FP)
    print("precision: ", precision)
    
    recall = TP / (TP+FN)
    print("recall: ", recall)
    
    
    f1_under = precision + recall
    f1_upper = precision * recall
    if f1_under == 0:
        print("precision+recall = 0")
        f1 = 0
    else:
        f1 = 2 * (f1_upper/f1_under)
    
    print("f1: ", f1)
```
