```python
import pandas as pd
df = pd.read_csv('./full_dataset.csv')
```

```python
df
```

```python
df = df.drop(['ResponseTimeTimeMedian','ResponseTimeTimeSkewFromMedian'],axis=1)
```

```python
X = df.drop(['Label'], axis=1)
y = df['Label']
```

```python
X
```

```python
y
```

Data preprocessing

```python
X.isnull().sum()
```

```python
X=X.dropna()
```

Dataset split

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X)

# 학습 데이터 스케일링
X_scaled = scaler.transform(X)
X = X_scaled
```

```python
X_5 = X[[0,1,2,3,4]]
X_5
```

## Pearson correlation

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
cor = X_5.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```
